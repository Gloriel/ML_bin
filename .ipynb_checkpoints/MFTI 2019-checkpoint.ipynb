```python
class Dragon:
    def __init__(self, name):
        self.name = name
        self.health = 100

    def is_alive(self):
        return self.health > 0

    def get_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0

    def talk(self):
        print(self.name, ': my health - ', self.health, 'Hit me!')

    def final_cry(self):
        print(self.name, 'died')
        
def main():
    enemy_list = [Dragon('Smoge'), Dragon('Hydra')]
    finish = False
    while not finish:
        dragon = enemy_list[0]
        dragon.talk()
        damage = int(input())
        dragon.get_damage(damage)
        if not dragon.is_alive():
            dragon.final_cry()
            enemy_list.pop(0)
        if not enemy_list:
            finish = True
            print('You WON!!')

main()
```

```python
#Защита от дурака "Сет-гет". позволяет ограничить возможность работы с хранимым числом. Не может, ufпример стать отрицательным 

class PositiveInt:
    
    __a = 0
    __counter = 0
    
    def set_a(self, a):
        self.__counter +=1
        if a >=0:
            self.__a = int(a)
        else:
            print("Wrong parameter, an internal state won't change")
            
    def get_a(self):
        print("Was set", self.__counter, "times.")
        return self.__a


if __name__ == "__main__":
    value = PositiveInt()
    
    print(value.get_a())
    
    value.set_a(10)
    print(value.get_a())
    
    value.set_a(-10)
    print(value.get_a())    
```

```python
#Waterfall - принцип последовательной разработки от проектирования до релиза
#Результат работы QA - повышение качества программного продукта (системное, модульное, интеграционное тестирование)
#Параметры качества: функциональеость, надежность, практичность, эффективность, сопровождаемость, переносимость
#Unit testing - модульное тестирование

def fib(n):
    if not isinstance(n, int): #Проверка на целое число
        raise TypeError("Fibonacci can work only with integer type")
    if n < 0: #проверка на отрицательные числа
        raise ValueError("Fibonacci can't work with negative numbers")
    if n >= 10000:
        raise ValueError("Fibonacci can't work with numbers larger than 10 000")
    F = [0, 1] + [0]*(n-1) #Добавляет в список n элементов
    for i in range(2, n+1):
        F[i] = F[i-1] + F[i-2]
    return F[n]

#Рефакторинг fib()

def fib(n):
    if not isinstance(n, int): #Проверка на целое число
        raise TypeError("Fibonacci can work only with integer type")
    if n < 0: #проверка на отрицательные числа
        raise ValueError("Fibonacci can't work with negative numbers")
    if n >= 10000:
        raise ValueError("Fibonacci can't work with numbers larger than 10 000")
    if n == 0:
        return 0
    f_2 = 0
    f_1 = 1 
    for i in range(2, n+1):
        f_1, f_2 = (f_1 + f_2), f_1 #Множественное присваивание 
    return f_1
   

#Программа по поиску числа фибоначчи для пользователя   
def main():
    n = int(input("Введите номер числа Фибонначи: "))
    f = fib(n)
    print("ваше число Фибоначчи: ", f)
    
#Опережающее тестирование. TDD,  Тест пишется заранее, на стадии проектирования модуля программы
#Простое и стрессовое тестирование. Оценка времени вычислений. Краш-тест.

all_correct = True

for n, answer in [(0,0), (1,1), (2,1), (5,5)]: 
    result = fib(n)
    correct = (result == answer)
    if not correct:
        print('Testing fibonacci() failure: ', n, "!=", answer)
    all_correct &= correct #Проверка на истинность утверждения

if all_correct:
    print('Testing fibonacci(): OK')
else:
    print('Testing fibonacci(): FAILED')
    

    
main()    
```
